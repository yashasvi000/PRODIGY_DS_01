```python
import pandas as pd
import matplotlib.pyplot as plt

# replace with the actual filename you uploaded
# The error message suggests that the CSV file has extra lines at the beginning.
# We can use the 'skiprows' parameter to skip these lines.
# Based on the error message "Expected 3 fields in line 5, saw 70", it seems the actual data starts after line 4.
df = pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv", skiprows=4)

# check first few rows
df.head()
```

```python
print(df.columns.tolist())
```

```python
# Select India and the indicator "Population, total"
india_df = df[(df['Country Name'] == "India") & (df['Indicator Name'] == "Population, total")]

# Keep only year columns (1960–2024)
years = [str(y) for y in range(1960, 2025)]
india_pop = india_df[years].T   # transpose so years become rows
india_pop.columns = ["Population"]
india_pop.index = india_pop.index.astype(int)  # convert index to int
india_pop.head()
```

```python
plt.figure(figsize=(10,6))
india_pop.tail(10).plot(kind="bar", legend=False, color="skyblue")
plt.title("India Population (2015–2024)")
plt.xlabel("Year")
plt.ylabel("Population")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(india_pop["Population"], bins=15, color="orange", edgecolor="black")
plt.title("Distribution of India’s Population (1960–2024)")
plt.xlabel("Population")
plt.ylabel("Frequency")
plt.show()
```
